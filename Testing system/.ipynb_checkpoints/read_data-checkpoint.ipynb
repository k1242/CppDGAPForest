```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Preparation

```python
def scale(x):
    return np.log(x)
#     return x / 1e6
```

```python
def prepare_data(x, y):
    data_x = []
    data_y = []
    data_num = []
    for (x1, y1) in zip(x, y):
        if x1 in data_x:
            data_y[data_x.index(x1)] += y1
            data_num[data_x.index(x1)] += 1
        else:
            data_x.append(x1)
            data_y.append(y1)
            data_num.append(1)
    x_ = np.array(data_x)
    y_ = np.array(data_y) / np.array(data_num)
    return x_, y_
```

```python
def hl_clip(x, y):
    tmp_data = pd.DataFrame({"x":x,"y":y})
    tmp_data = tmp_data[tmp_data.x>8]
    return np.array(tmp_data.x), np.array(tmp_data.y)
```

```python
def zero_x(x):
    return 0*x
```

```python
def show_graph(data_name, clip, name, shift=zero_x):
    data1 = pd.read_csv(data_name, names=["n", "time"])
    x1 = scale(data1.n)
    y1 = data1.time

    x = np.array(x1)
    y = np.array(y1).clip(0, clip)
    print("amount of points:\t", len(x))
    x, y = prepare_data(x, y)
    y = y - shift(x)
    print("amount of unique points:", len(x))

    plt.figure(dpi=300)
    plt.scatter(x, y, s=1, alpha=0.5, color="k")

    x2l, y2l = hl_clip(x, y)
    coef, cov = np.polyfit(x2l, y2l, 1, cov=True)
    title = "t = ({:.2f} +- {:.2f})ms log(N) + ({:.2f} +- {:.2f})ms".format(
        coef[0] * 1000, cov[0][0]**0.5 * 1000, 
        coef[1] * 1000, cov[1][1]**0.5 * 1000)
    
    print("\nпри N > 3000:\n" + title)
    f = np.poly1d(coef)

    x_ = np.linspace(0, max(x), 10)
    plt.plot(x_, f(x_), 
             linewidth=5, alpha=0.5,
             color="green", label="$\\mathrm{LS\\ approximation}$")

    plt.xlabel("$\\log\\ {N}$")
    plt.ylabel("$t, \\mathrm{ns}$")
    plt.title(name+": " + title)
    plt.legend()
    plt.xlim(0, 19)
    plt.ylim(0, max(y)*1.05)
    plt.savefig("figures\\" + data_name[:-4] + ".png")
    plt.show()
```

```python
def get_shift(data_name, clip, deg=3):
    data1 = pd.read_csv(data_name, names=["n", "time"])
    x1 = scale(data1.n)
    y1 = data1.time

    x = np.array(x1)
    y = np.array(y1).clip(0, clip)
    x, y = prepare_data(x, y)

    coef, cov = np.polyfit(x, y, deg, cov=True)
    f = np.poly1d(coef)
    return f
```

## Test AVL structure

```python
show_graph("AVL_remove.csv", name="AVL_remove", clip=0.2)
```

```python
# insert only
show_graph("AVL_inser_remove.csv", 
           clip = 0.3,
           name = "AVL_insert",
           shift = get_shift("AVL_remove.csv", clip=0.8))
```

```python
show_graph("AVL_find.csv", name = "AVL_find", clip=0.1)
```

## Test RB structure

```python
show_graph("RB_remove.csv", name="RB_remove", clip=0.6)
```

```python
show_graph("RB_inser_remove.csv", 
           clip=0.35,
           name="RB_insert",
           shift=get_shift("RB_remove.csv", clip=0.8))
```

```python
show_graph("RB_find.csv", name="RB_find", clip=0.9)
```

## Overview

```python
struct_names = ["AVL", "RB"]
plt.figure(dpi=300)
linestyle = "-"
deg = 7

for struct_name in struct_names:
    f_r  = get_shift(struct_name + "_remove.csv", 0.2, deg=deg)
    f_ir = get_shift(struct_name + "_inser_remove.csv", 0.3, deg=deg)
    f_f  = get_shift(struct_name + "_find.csv", 0.3, deg=deg)
    x_ = np.linspace(0, 18, 100)
    plt.plot(x_, f_r(x_), 
             color = "r", linestyle=linestyle,
             linewidth=3, alpha=0.1)
    plt.plot(x_, f_ir(x_)-f_r(x_), 
             color = "g", linestyle=linestyle,
             linewidth=3, alpha=0.1)
    plt.plot(x_, f_f(x_), 
             color = "b", linestyle=linestyle,
             linewidth=3, alpha=0.1)
    

    f_r  = get_shift(struct_name + "_remove.csv", 0.2, deg=1)
    f_ir = get_shift(struct_name + "_inser_remove.csv", 0.3, deg=1)
    f_f  = get_shift(struct_name + "_find.csv", 0.3, deg=1)
    x_ = np.linspace(8, 18, 100)
    plt.plot(x_, f_r(x_), 
             color = "r", linestyle=linestyle,
             linewidth=3, alpha=0.6, 
             label=struct_name+"_remove")
    plt.plot(x_, f_ir(x_)-f_r(x_), 
             color = "g", linestyle=linestyle,
             linewidth=3, alpha=0.6, 
             label=struct_name+"_insert")
    plt.plot(x_, f_f(x_), 
             color = "b", linestyle=linestyle,
             linewidth=3, alpha=0.6, 
             label=struct_name+"_find")
    linestyle = "--"
plt.legend()
plt.savefig("figures/overview.png")
plt.show()
```
